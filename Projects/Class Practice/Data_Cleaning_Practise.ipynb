```python
from scipy import stats

# Calculate z-scores for the column
z_scores = stats.zscore(df['num_critic_for_reviews'])

# Define a threshold for outlier detection
threshold = 3

# Remove rows with z-scores above the threshold
df = df[(z_scores < threshold)]
```

```python
from scipy import stats

# Calculate z-scores for the column
z_scores = stats.zscore(df['duration'])

# Define a threshold for outlier detection
threshold = 3

# Remove rows with z-scores above the threshold
df = df[(z_scores < threshold)]
```

```python
mean_value = df['num_critic_for_reviews'].mean()
df.loc[df['num_critic_for_reviews'] > threshold, 'num_critic_for_reviews'] = mean_value
```

```python
mean_value = df['duration'].mean()
df.loc[df['duration'] > threshold, 'duration'] = mean_value
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Apply feature scaling to a specific column
df['num_critic_for_reviews'] = scaler.fit_transform(df['num_critic_for_reviews'])
```
