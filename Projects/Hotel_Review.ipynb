```python
# Select specific columns from the existing DataFrame
new_df = df[['Hotel_Address', 'Negative_Review', 'Positive_Review', 'Reviewer_Score']].copy()

# Verify the new DataFrame
(new_df.head())
```

1) Merge Negative Review and Positive Review Column as One Column named Review

```python
# Remove rows with 'No Negative' and 'No Positive' reviews
new_df = new_df[~((new_df['Negative_Review'] == 'No Negative') & (new_df['Positive_Review'] == 'No Positive'))].copy()

# Reset the index of the new DataFrame
new_df.reset_index(drop=True, inplace=True)

# Continue with the remaining steps

# Select specific columns from the existing DataFrame
exploratoryanalysisdf = new_df[['Hotel_Address', 'Negative_Review', 'Positive_Review', 'Reviewer_Score']].copy()

# Merge 'Negative_Review' and 'Positive_Review' columns
exploratoryanalysisdf['review'] = exploratoryanalysisdf['Negative_Review'].str.cat(exploratoryanalysisdf['Positive_Review'], sep=' ')

# Count the number of words in 'Negative_Review' and 'Positive_Review' columns
exploratoryanalysisdf['Negative Review Word Count'] = exploratoryanalysisdf['Negative_Review'].str.split().apply(len)
exploratoryanalysisdf['Positive Review Word Count'] = exploratoryanalysisdf['Positive_Review'].str.split().apply(len)

# Count the number of reviews for each property
property_reviews_count = exploratoryanalysisdf['Hotel_Address'].value_counts()

# Create a dictionary mapping properties to review counts
property_to_reviews = {
    address: count for address, count in zip(property_reviews_count.index, property_reviews_count.values)
}

# Add 'Total_Reviews' column to exploratoryanalysisdf
exploratoryanalysisdf['Total_Reviews'] = exploratoryanalysisdf['Hotel_Address'].map(property_to_reviews)

# Verify the updated DataFrame
print(exploratoryanalysisdf.head())
```
