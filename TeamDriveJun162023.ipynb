```python
import pandas as pd

pathSales = '/content/gdirve/MyDrive/Colab Notebooks/TakeoData/sales.csv'
dfSales = pd.read_csv(pathSales)
```

```python
dfSales
```

```python
pathStores = '/content/gdirve/MyDrive/Colab Notebooks/TakeoData/stores.csv'
dfStores = pd.read_csv(pathStores)
dfStores
```

```python
# Merge the datasets based on a common column
merged_df = pd.merge(dfSales, dfStores, on='store_nbr')
```

```python
# Group by store_nbr and calculate the sum of sales
grouped_df = merged_df.groupby('store_nbr')['sales'].sum()
```

```python
import matplotlib.pyplot as plt

# Plot sales by store_nbr
plt.bar(grouped_df.index, grouped_df)
plt.xlabel('Store Number')
plt.ylabel('Total Sales')
plt.title('Sales by Store Number')
plt.show()
```

```python
# Calculate descriptive statistics for sales in dataset 1
df1_stats = dfSales['sales'].describe()

# Calculate descriptive statistics for sales in dataset 2
df2_stats = dfStores['sales'].describe()

# Print the statistics
print("Dataset 1 Statistics:\n", df1_stats)
print("\nDataset 2 Statistics:\n", df2_stats)
```
