```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
path = '/content/Salary_Data.csv'
df = pd.read_csv(path)
df
```

```python
#separating data into dependent and independent variables

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#model making

model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
#prediction from model after successfully constructing the model

#for_which = int(input('Enter the yearly experience for which you want to estimate the salary for: '))
salary = model.predict([[2],[1],[8]])

print('The estimated salary is ', salary)
```

```python
path = '/content/50_Startups.csv'
df2 = pd.read_csv(path)
df2
```

```python
# Perform one-hot encoding
df2 = pd.get_dummies(df2, columns=['State'])
```

```python
df2
```

```python
from collections.abc import ValuesView
#separating data into dependent and independent variables

X2 = df2.iloc[:, :].values
y2 = df2['Profit'].values

#because 'Profit' column is in middle of the DF, we drop it for the consistency of independent values
X2 = df2.drop(['Profit'], axis = 1).values


#as a normal pracitise, we must drop any columns after assigning the dependent/independent variables
```

```python
#model making

model2 = linear_model.LinearRegression()
model2.fit(X2, y2)
```

```python
#for_which = int(input('Enter the yearly experience for which you want to estimate the salary for: '))
#R&D Spend	Administration	Marketing Spend	Profit	State_California	State_Florida	State_New York

rd = float(input('Enter R&D amount: '))
admin = float(input('Enter Administration amount: '))
market = float(input('Enter Markeing amount: '))
state = int(input('For California, Enter 1; For Florida, Enter 2; For New York, Enter 3; '))
cali, flo, ny = 0,0,0
if (state == 1):
  cali = 1
  print('The state selected is California')
elif (state == 2):
  flo = 1
  print('The state selected is Florida')
elif (state == 3):
  ny = 1
  print('The state selected is New York')
else:
  print('Invalid state choice, program will crash')

prediction = model2.predict([[rd, admin, market, cali, flo, ny]])

print('The predicted Profit is ', prediction)
```
