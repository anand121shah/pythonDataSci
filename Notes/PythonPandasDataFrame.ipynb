#**Test Case 4**

```python
# importing important library

import pandas as pd
```

```python
df_users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
df_users.head()
```

```python
df_users.groupby('occupation').age.mean()
```

```python
# create a function
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

# apply the function to the gender column and create a new column
df_users['gender_n'] = df_users['gender'].apply(gender_to_numeric)


a = df_users.groupby('occupation').gender_n.sum() / df_users.occupation.value_counts() * 100 

# sort to the most male 
a.sort_values(ascending = False)
```

```python
df_users.groupby('occupation').age.agg(['min', 'max'])
```

```python
df_users.groupby(['occupation', 'gender']).age.mean()
```

```python
# create a data frame and apply count to gender
gender_ocup = df_users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

# create a DataFrame and apply count for each occupation
occup_count = df_users.groupby(['occupation']).agg('count')

# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100

# present all rows from the 'gender column'
occup_gender.loc[: , 'gender']
```
