```python
#DATA CLEANING
#DROPPING DUPLICATES
data = df.dropna().copy()

data['Minutes'] = data['Minutes'].str.split(' ').str[0]
data.Minutes = data.Minutes.astype(int)

data['Votes'] = data['Votes'].str.replace(',', '')
data.Votes = data.Votes.astype(int)
```

```python
label = data["genre"].value_counts().index
sizes = data["genre"].value_counts()
plt.figure(figsize = (16,16))
plt.pie(sizes, labels= label, startangle = 0 , shadow = False , autopct='%1.1f%%')
plt.show()
```

```python
order_rate = data.groupby('Name').mean().sort_values('Rating', ascending = False).index

plt.figure(figsize=(20, 12))
ax = sns.barplot(y='Name', x='Rating', data = data, order=order_rate)
ax.bar_label(ax.containers[0])
plt.title('Movie/TV Show with highest rating')
plt.show()
plt.clf()
```

```python
order_min = data.groupby('Name').mean().sort_values('Minutes', ascending = False).index

plt.figure(figsize=(20, 12))
ax = sns.barplot(y='Name', x='Minutes', data = data, order=order_min)
ax.bar_label(ax.containers[0])
plt.title('Movie/TV Show with highest Minutes')
plt.show()
plt.clf()
```

```python
order_votes = data.groupby('Name').mean().sort_values('Votes', ascending = False).index

plt.figure(figsize=(20, 12))
ax = sns.barplot(y='Name', x='Votes', data = data, order=order_votes)
ax.bar_label(ax.containers[0])
plt.title('Movie/TV Show with highest votes')
plt.show()
plt.clf()
```

```python
px.scatter(data,x='Rating',y='Votes',color='Rating',color_continuous_scale=px.colors.sequential.Plasma,template='plotly_dark',title='<b>Rating Versus Votes')
```
