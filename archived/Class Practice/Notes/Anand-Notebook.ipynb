```python
#Write a program to print twin primes less than 1000. 
#If two consecutive odd numbers are both prime then they are known as twin primes output 3,5   5,7  11,13

oldprime = 3
newprime = 5

rangeNum = 1000

for k in range(5, rangeNum):

  #checking if current value of k is prime
  x = True
  for i in range(2, k):
    if k%i == 0:
      x = False

  #this part will not execute until current value of k is prime
  if x == True:
    if newprime-oldprime == 2:
      print(oldprime, newprime)
    oldprime = newprime
    newprime = k
```

```python
#If all digits of a number n are multiplied by each other repeating with the product, 
#the one digit number obtained at last is called the multiplicative digital root of n.
#The number of times digits need to be multiplied to reach one digit is called the multiplicative persistance of n

#Example: 86 -> 48 -> 32 -> 6 (MDR 6, MPersistence 3)
# 341 -> 12->2           (MDR 2, MPersistence 2)


def selfMulti(n):
  a = 1
  for i in range(len(str(n))):
    mod = n%10
    a = a*mod
    n = n//10
  return a

givenNum = int(input("Enter a number: "))
print("The entered number is:", givenNum)
MDR = givenNum

count = 0
while(MDR>9):
  MDR = selfMulti(MDR)
  count = count+1

print('MDR is', MDR)
print('MPersistence is', count)
```

```python
# super() function

class Alpha:
	classvar1 = "classvar1 - Class Alpha - Class local variable"
	def __init__(self):
		self.var1 = "self.var1 - Class Alpha - Constructor or __init__ function"
		self.classvar1 = "self.classvar1 - Class Alpha - Instance variable"

class Beta(Alpha):
  classvar1 = "Class Beta - Class local variable"
  def __init__(self):
    
    self.var1 = "self.var1 - Class Beta - Constructor or __init__ function"
    self.classvar1 = "self.classvar1 - Class Beta - Instance variable"
    super().__init__()
a = Alpha()
b = Beta()

print(b.classvar1, b.var1)
```

```python
# Method Overriding

class Alpha:
	classvar1 = "classvar1 - Class Alpha - Class local variable"
	def __init__(self):
		self.var1 = "self.var1 - Class Alpha - Constructor or __init__ function"
		self.classvar1 = "self.classvar1 - Class Alpha - Instance variable"

class Beta(Alpha):
  classvar1 = "Class Beta - Class local variable"
  def __init__(self):
    self.var1 = "self.var1 - Class Beta - Constructor or __init__ function"
    self.classvar1 = "self.classvar1 - Class Beta - Instance variable"
a = Alpha()
b = Beta()

print(b.var1, b.classvar1)
```

```python
#a class to show example of overriding using super function

class Drive:
    driver = 'Default Driver Input'

    def __init__(self, driver):
        self.driver = 'default'

    def acc(self, speed):
        print(self.driver, 'is driving at', 'speed')

class Autopilot(Drive):
    driver = 'Default Autopilot Driver'

    def __init__(self, driver):
        super().__init__(driver)
        self.autopilot = 'Autopilot Driver'

    def acc(self, speed):
        speed = 60
        print(self.autopilot, 'is driving at', speed)

class Manual(Drive):
    driver = 'Human Driver'

    def __init__(self, humanName):
        super().__init__(humanName)
        self.humanName = humanName + ' Driver'

    def acc(self, speed):
        print(self.humanName, 'is driving at', speed)

# Creating instances of the classes and invoking the overridden methods

drive1 = Drive('John')
drive1.acc(50)  # Output: John is driving at 50

autopilot1 = Autopilot('Tesla Autopilot')
autopilot1.acc(0)  # Output: Autopilot Driving is driving at 60

manual1 = Manual('Sarah')
manual1.acc(40)  # Output: Sarah Driver is driving at 40
```

```python
# define Python user-defined exceptions
#exception:

#1. UDExc: valid age > 19
#2. UDExc: gender in ['M', 'F']
#3. UDExc: if 'M' >= 160 cms
#        if 'F' >= 150 cms
#4. UDExc: if 'M' <= 72 and >= 65
#        if 'F' <= 65 and >= 55
#5. UDExc: Nationality = 'INDIAN'

class InvalidAgeException(Exception):
    pass

class InvalidGenderException(Exception):
    pass

class InvalidHeightException(Exception):
    pass

class InvalidWeightException(Exception):
    pass

class InvalidNationalityException(Exception):
    pass

input_age = int(input("Enter an age in years: "))
input_gen = input("Enter a gender ('M' or 'F'): ").upper()
input_height = int(input("Enter a height in CMs: "))
input_weight = int(input("Enter a weight in KGs: "))
input_nationality = input("Enter a nationality according to your passport: ").upper()

try:
    if input_age <= 19:
        raise InvalidAgeException("Invalid age. Age should be greater than 19.")
    if input_gen not in ['M', 'F']:
        raise InvalidGenderException("Invalid gender. Gender should be 'M' or 'F'.")
    if input_gen == 'M' and not (160 <= input_height):
        raise InvalidHeightException("Invalid height for male. Height should be greater than or equal to 160 cms.")
    if input_gen == 'F' and not (150 <= input_height):
        raise InvalidHeightException("Invalid height for female. Height should be greater than or equal to 150 cms.")
    if input_gen == 'M' and not (65 <= input_weight <= 72):
        raise InvalidWeightException("Invalid weight for male. Weight should be between 65 and 72 kgs.")
    if input_gen == 'F' and not (55 <= input_weight <= 65):
        raise InvalidWeightException("Invalid weight for female. Weight should be between 55 and 65 kgs.")
    if input_nationality != 'INDIAN':
        raise InvalidNationalityException("Invalid nationality. Nationality should be 'INDIAN'.")
    
    # If all conditions pass, the candidate is valid
    if input_gen == 'M':
        print('Congrats, you are Mr. India')
    elif input_gen == 'F':
        print('Congrats, you are Ms. India')
except (InvalidAgeException, InvalidGenderException, InvalidHeightException, InvalidWeightException, InvalidNationalityException) as e:
    print("Exception occurred:", str(e))
```

```python
#Write a Python class Employee with attributes like emp_id, emp_name, emp_salary, and emp_department and methods like calculate_emp_salary, emp_assign_department, and print_employee_details.
#Use 'assign_department' method to change the department of an employee.
#Use 'print_employee_details' method to print the details of an employee.
#Use 'calculate_emp_salary' method takes two arguments: salary and hours_worked, which is the number of hours worked by the employee.

class Employee:
  def __init__(self, emp_id, emp_name, emp_salary, emp_department):
    self.emp_id = emp_id
    self.emp_name = emp_name
    self.emp_salary = emp_salary
    self.emp_department = emp_department

  def assign_department(self, new_department):
    self.emp_department = new_department
    print('The department of ', self.emp_name, 'is', self.emp_department)

  def print_employee_details(self):
    print('The employee name is',self.emp_name)
    print('The employee ID is', self.emp_id)
    print('The employee salary in dollars per hour is', self.emp_salary)
    print('The employee department is', self.emp_department)
  
  def calculate_emp_salary(self, hours_worked):
    print('Total Salary is ', self.emp_salary*hours_worked)
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Create a colorful spiral pattern
theta = np.linspace(0, 10 * np.pi, 1000)
radius = np.linspace(0.1, 1, 1000)
colors = np.linspace(0, 1, 1000)
x = radius * np.cos(theta)
y = radius * np.sin(theta)

#Plot the spiral with changing colors
plt.scatter(x, y, c=colors, cmap='rainbow')
plt.colorbar(label='Color')
plt.title('Colorful Spiral Pattern')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

#Generate random data points
np.random.seed(42)
x_values = np.random.randn(100)
y_values = np.random.randn(100)
colors = np.random.randn(100)
sizes = np.random.randint(10, 200, 100)

#Plot the random data points with varying colors and sizes
plt.scatter(x_values, y_values, c=colors, s=sizes, cmap='coolwarm')
plt.colorbar(label='Magnitude')
plt.title('Random Data Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

#Create a pie chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [30, 15, 20, 10, 25]
explode = [0, 0.1, 0, 0, 0]
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']

plt.pie(values, labels=categories, explode=explode, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Distribution of Categories')
plt.axis('equal')
plt.show()
```

```python
# WAP in which if file does not exists you will create a csv file with 1 column and add 1 row wih value 1

# and if file exists read the column add 1 to the last value of the column and add the new value to csv and save

# import os

# if os.path.exists("abcd.csv")

import os
import pandas as pd

file_path = "abcd.csv"

# Check if the file exists
if os.path.exists(file_path):
    # Read the existing file
    df = pd.read_csv(file_path)
    
    # Get the last value of the column
    last_value = df.iloc[-1, 0]
    
    # Increment the value by 1
    new_value = last_value + 1
    
    # Add the new value as a new row
    new_row = pd.DataFrame({"Column1": [new_value]})
    df = df.append(new_row, ignore_index=True)
else:
    # Create a new dataframe with one column and one row
    df = pd.DataFrame({"Column1": [1]})

# Save the dataframe to a CSV file
df.to_csv(file_path, index=False)
```
